### leetcode 180 

이 문제는 3개 이상의 연속된 수를 뽑아내는 문제이다. 

### IDEA

아이디어는 윈도우 함수인 lead, lag 을 사용하여 

`이전 값과 이후 값을 계산한 컬럼을 만들고, num 컬럼과 prev, aft 컬럼의 값이 같다면`

`3개 이상의 연속된 수가 나오지 않을까 하는 생각으로 작성.`

`lead 는 이후 값, lag 는 이전 값을 계산한다.`

| id | num | prev | aft |
|----|-----|------|-----|
| 1  |  1  |  n   |  1  |
| 2  |  1  |  1   |  1  | 
| 3  |  1  |  1   |  2  |
| 4  |  2  |  1   |  1  |
| 5  |  1  |  2   |  2  |
| 6  |  2  |  1   |  2  |
| 7  |  2  |  2   |  n  |

위의 표와 같이 num, prev, aft 의 값이 모두 같다면, 3개 이상의 컬럼이 같은 값인 것을 알 수 있다.

### 쿼리

```python
select distinct num as ConsecutiveNums

from (select num,
      lag(num) over (order by id asc) prev_num      
      lead(num) over (order by id asc) aft_num,
      from logs) as sub_table
      
where num = prev_num and num = aft_num
```

### 삽질

```python
select distinct ConsecutiveNums

from (select 
      id,
      num as ConsecutiveNums,
      row_number() over (order by num) as row_num,
      (id - num + 1) as checker
      from logs) as sub_table

where row_num = checker
```

위의 쿼리와 같이 이상한 짓도 많이했다.

1. num 이 음수인 경우에 예외 case

2. 3번 이상의 조건을 뽑아낼 방법을 못찾음.

   ㄴ group by 로 묶고 having count 로 뽑아내기도 안됨.
